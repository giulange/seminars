#### PARS

```matlab
WDIR        = '~/git/seminars/Aedes-Procida/';
OVITRAPS    = 'analisi_spaziale_luglio_settembre_2016.xls';% --> copied from Dropbox/dati_finali_procida
MAT_FILE    = 'data.mat';
FIL_dem_32632 = fullfile('~/work/People/Michela/andosols_spatstat','dem_20m_ispra','dtm_20m_32.tif');   %  20m DEM, EPSG:32632
FIL_dtm     = '~/Dropbox/dati_finali_procida/dati/DTM.tif';
```

#### Ovitraps

```matlab
[~,~,OVI] = xlsread( fullfile(WDIR,OVITRAPS), 'SPAZIALE_Giuliano' );

%head(OVI)

OVI = OVI(:,[1,2,19,20,5:8,4]);

head(OVI)
```

#### -- convert (Lat,Lon) to EPSG::32632

```matlab
%proj = geotiffinfo(FIL_dem_32632);
%[Easting, Northing] = projfwd( proj, cae(OVI,3), cae(OVI,4) );
```

#### -- convert (Lat,Lon) to EPSG::32633

```matlab
proj = geotiffinfo(FIL_dtm);
[Easting, Northing] = projfwd( proj, cae(OVI,3), cae(OVI,4) );
```

#### ---- write EPSG::32633

```matlab
OVI(1,size(OVI,2)+1)={'Easting'};
OVI(2:end,size(OVI,2))=num2cell(Easting);
OVI(1,size(OVI,2)+1)={'Northing'};
OVI(2:end,size(OVI,2))=num2cell(Northing);
```

#### -- retrieve orthophoto

```matlab
% ***GET ORTHOPHOTO
Url_ortofoto = 'http://sit.regione.campania.it:80/arcgis/services/cartografia_base/MapServer/WMSServer';
ortofoto = wmsinfo(Url_ortofoto);

LatProcida = [min(cae(OVI,3)),max(cae(OVI,3))];
LonProcida = [min(cae(OVI,4)),max(cae(OVI,4))];
% to retireve the orthophoto:
LatE = [LatProcida(1)-0.005, LatProcida(2)+0.005];
LonE = [LonProcida(1)-0.005, LonProcida(2)+0.005];
% [LatE,LonE]
proj = geotiffinfo(FIL_dtm);
[Ebox, Nbox] = projfwd( proj, LatE, LonE );
Eline=[Ebox(1),Ebox(1),Ebox(2),Ebox(2),Ebox(1)];
Nline=[Nbox(1),Nbox(2),Nbox(2),Nbox(1),Nbox(1)];

figure(1),clf
mapshow(cae(OVI,10),cae(OVI,11),'DisplayType','point','Marker','o',...
        'MarkerEdgeColor', 'k','MarkerFaceColor','r','MarkerSize',8)
hold on
mapshow(Eline,Nline,'DisplayType','line','Marker','+',...
        'MarkerEdgeColor', 'k','MarkerFaceColor','r','MarkerSize',8)
hold off

[sg,Rsg] = wmsread(ortofoto.Layer(1),'Latlim',LatE,'Lonlim',LonE);%,'ImageFormat','image/tiff');
% Display the ortho-image in a UTM projection.
figure(2), clf
axesm('utm', 'Zone', utmzone(LatE, LonE), ...
    'MapLatlimit', LatE, 'MapLonlimit', LonE, ...
    'Geoid', wgs84Ellipsoid)
geoshow(sg,Rsg)
axis off
title({'\fontsize{14}\color{black}\bf Procida Island','\fontsize{12}\color{red}Orthophoto & point monitoring','\fontsize{8}\color{gray}Aedes albopictus'})
hold on
geoshow(cae(OVI,3),cae(OVI,4),'DisplayType','point','Marker','o',...
        'MarkerEdgeColor', 'k','MarkerFaceColor','r','MarkerSize',4)
hold off
% axis equal
```

#### -- [ISSUE] retrieve 1m Lidar Campania Region

```matlab
url_lidar='http://wms.pcn.minambiente.it/ogc?map=/ms_ogc/WMS_v1.3/servizi-LiDAR/LIDAR_CAMPANIA.map';
lidar = wmsinfo(url_lidar);

LatProcida = [min(cae(OVI,3)),max(cae(OVI,3))];
LonProcida = [min(cae(OVI,4)),max(cae(OVI,4))];
% to retireve the orthophoto:
LatE = [LatProcida(1)-0.005, LatProcida(2)+0.005];
LonE = [LonProcida(1)-0.005, LonProcida(2)+0.005];

% scrutando gli 8 layers, evinco che il numero 3 potrebbe fare al caso
% nostro:
[Lidar,Rlidar] = wmsread(lidar.Layer(3),'Latlim',LatE,'Lonlim',LonE);%,'ImageFormat','image/tiff');

figure(3), clf
axesm('utm', 'Zone', utmzone(LatE, LonE), ...
    'MapLatlimit', LatE, 'MapLonlimit', LonE, ...
    'Geoid', wgs84Ellipsoid)
geoshow(Lidar,Rlidar)
```

#### -- [ISSUE] retrive 2m/4m Lidar-based DTM

```matlab
url_dem_camp = 'http://wms.pcn.minambiente.it/ogc?map=/ms_ogc/WMS_v1.3/servizi-LiDAR/AMP_LiDAR_Multibeam_Campania.map';
lidar4m = wmsinfo(url_dem_camp);

LatProcida = [min(cae(OVI,3)),max(cae(OVI,3))];
LonProcida = [min(cae(OVI,4)),max(cae(OVI,4))];
% to retireve the orthophoto:
LatE = [LatProcida(1)-0.005, LatProcida(2)+0.005];
LonE = [LonProcida(1)-0.005, LonProcida(2)+0.005];

% scrutando gli 8 layers, evinco che il numero 3 potrebbe fare al caso
% nostro:
[Lidar4m,Rlidar4m] = wmsread(lidar4m.Layer(3),'Latlim',LatE,'Lonlim',LonE,'ImageFormat','image/tiff');

figure(3), clf
axesm('utm', 'Zone', utmzone(LatE, LonE), ...
    'MapLatlimit', LatE, 'MapLonlimit', LonE, ...
    'Geoid', wgs84Ellipsoid)
geoshow(Lidar4m,Rlidar4m)
```

#### DEM & derivatives

#### Resample dtm to 10m

```matlab
info = geotiffinfo(FIL_dtm);
[dtm2,Rdtm]=geotiffread(FIL_dtm);
dtm2(dtm2==-9999)=NaN;
[dtm10,Rdtm10] = resamplem(dtm2,info.RefMatrix, 10, 'bilinear', NaN);
[dtm20,Rdtm20] = resamplem(dtm2,info.RefMatrix, 20, 'bilinear', NaN);
```

```matlab
FIL_dtm_10m = fullfile(WDIR,'dtm10m');
geotiffwrite(FIL_dtm_10m,dtm10,Rdtm10, 'GeoKeyDirectoryTag', info.GeoTIFFTags.GeoKeyDirectoryTag)
```

```matlab
FIL_dtm_20m = fullfile(WDIR,'dtm20m');
geotiffwrite(FIL_dtm_20m,dtm20,Rdtm20, 'GeoKeyDirectoryTag', info.GeoTIFFTags.GeoKeyDirectoryTag)
```
